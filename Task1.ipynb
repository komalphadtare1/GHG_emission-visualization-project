```python
df = df.dropna(subset=['Supply Chain Emission Factors with Margins'])
```

```python
df.shape
```

```python
df = df.drop_duplicates()
df.shape
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
top10 = df.groupby("Commodity Name")["Supply Chain Emission Factors with Margins"] \
          .sum().sort_values(ascending=False).head(10)
sns.barplot(x=top10.values, y=top10.index)
plt.title("Top 10 Industries with Highest Emissions")
plt.xlabel("Emission Value")
plt.ylabel("Industry")
plt.show()
```

```python
sns.histplot(df["Supply Chain Emission Factors with Margins"], bins=30, kde=True)
plt.title("Emission Value Distribution")
plt.xlabel("Emission")
plt.show()
```
